```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "title": [
        "Inception", "Titanic", "Avatar", "The Dark Knight",
        "Interstellar", "Joker", "Frozen", "Gladiator",
        "Avengers", "Coco", "Parasite", "Up", "The Lion King",
        "La La Land", "The Godfather"
    ],
    "genre": [
        "Sci-Fi", "Romance", "Sci-Fi", "Action",
        "Sci-Fi", "Drama", "Animation", "Action",
        "Action", "Animation", "Thriller", "Animation",
        "Animation", "Musical", "Crime"
    ],
    "rating": [
        8.8, 7.9, 7.8, 9.0,
        8.6, 8.4, 7.5, 8.5,
        8.0, 8.4, 8.6, 8.2,
        8.5, 8.0, 9.2
    ]
}

df = pd.DataFrame(data)
df
```

```python
# Check structure
df.info()

# Check missing values
df.isnull().sum()

# Remove duplicates (if any)
df.drop_duplicates(inplace=True)
```

```python
mean_rating = df["rating"].mean()
median_rating = df["rating"].median()
mode_rating = df["rating"].mode()[0]

print("Mean Rating:", mean_rating)
print("Median Rating:", median_rating)
print("Mode Rating:", mode_rating)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["rating"], bins=8, kde=True)
plt.title("Distribution of Movie Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["rating"])
plt.title("Boxplot of Movie Ratings")
plt.show()
```

```python
top_movies = df.sort_values(by="rating", ascending=False).head(5)
top_movies
```

```python
C = df["rating"].mean()
m = df["votes"].quantile(0.60)

def weighted_rating(x, m=m, C=C):
    v = x["votes"]
    R = x["rating"]
    return (v/(v+m) * R) + (m/(v+m) * C)

df["weighted_rating"] = df.apply(weighted_rating, axis=1)

df.sort_values("weighted_rating", ascending=False).head(5)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
np.random.seed(42)

titles = [f"Movie_{i}" for i in range(1, 201)]
genres = ["Action", "Drama", "Comedy", "Sci-Fi", "Thriller", "Animation", "Crime", "Romance"]

data = {
    "title": titles,
    "genre": np.random.choice(genres, 200),
    "year": np.random.randint(1980, 2024, 200),
    "rating": np.round(np.random.normal(7.5, 0.8, 200), 1),
    "votes": np.random.randint(50000, 3000000, 200),
    "budget_million": np.random.randint(10, 300, 200),
    "revenue_million": np.random.randint(20, 1000, 200)
}

df = pd.DataFrame(data)

# Clip ratings between 1 and 10
df["rating"] = df["rating"].clip(1, 10)

df.head()
```

```python
# Check missing values
df.isnull().sum()

# Remove duplicates
df.drop_duplicates(inplace=True)

# Ensure correct data types
df.dtypes
```

```python
df["profit_million"] = df["revenue_million"] - df["budget_million"]
```

```python
df["roi"] = (df["profit_million"] / df["budget_million"]) * 100
```

```python
df["rating_category"] = pd.cut(
    df["rating"],
    bins=[0, 6, 8, 10],
    labels=["Low", "Average", "High"]
)
```

```python
C = df["rating"].mean()
m = df["votes"].quantile(0.75)

def weighted_rating(x, m=m, C=C):
    v = x["votes"]
    R = x["rating"]
    return (v/(v+m) * R) + (m/(v+m) * C)

df["weighted_rating"] = df.apply(weighted_rating, axis=1)

top_movies = df.sort_values("weighted_rating", ascending=False).head(10)
top_movies[["title", "genre", "rating", "weighted_rating"]]
```
